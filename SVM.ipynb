```python
def svm(X,y):
    weights= np.zeros(X.shape[1])
    alpha=10
    epochs = 1000
    for epoch in range(1,epochs):
        for i,x in enumerate(X):
            if  ( y[i]  *  np.dot(X[i],weights)) <  1:
                weights=weights +alpha*((X[i]*y[i])+(-2 *(1/epoch)*weights))
            else:
                weights = weights+alpha *(-2   *(1/epoch)*  weights)
    return weights
```

```python
weights = svm (X,y)
```

```python
w= weights
```
